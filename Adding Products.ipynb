```python
import json
```

```python
fd = open("record.json","r")
r = fd.read()
fd.close()
```

```python
record = json.loads(r)
record
```

```python
prod_id = str(input("Enter product ID: "))
# checking if the product is already part of the inventory
if prod_id in record.keys():
    # modifying product info if it already exists
    ch = int(input(print("The product is already in the inventory\nEnter 1 to update the price\nEnter 2 to update the quantity\nEnter 3 to remove the product")))
    if (ch == 1):
        # updating price
        npr = int(input(print("Enter the new price : ")))
        record[prod_id]["pr"] = npr
    elif (ch == 2):
        # updating quantity
        nqn = int(input(print("Enter the new quantity : ")))
        record[prod_id]["qn"] = nqn
    elif (ch == 3):
        # removing products
        del record[prod_id]
    else:
        print("Invalid Input")
else:
    name = str(input("Enter name: "))
    pr = int(input("Enter price: "))
    qn = int(input("Enter quantity: "))
    s_brand = str(input("Enter brand name: "))
    s_type = str(input("Enter type: ")) 
    record[prod_id] = {'name': name, 'pr': pr, 'qn': qn, 's_brand': s_brand, 's_type': s_type}

js = json.dumps(record, indent=2, sort_keys=True)
f = open("record.json", 'w')
f.write(js)
f.close()
f = open("record.json", 'r')
r = f.read()
f.close()
print(r)
```
